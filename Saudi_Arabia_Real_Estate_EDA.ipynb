# **Handling Outlier**

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
```

```python
df_clean = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
print("Original shape:", df.shape)
print("New shape after removing outliers:", df_clean.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10,5))
sns.boxplot(x=df_clean['price'], color='skyblue')
plt.title("Boxplot of Property Prices (After Removing Outliers)")
plt.xlabel("Price")
plt.show()
```

```python
min_price = df_clean['price'].min()
max_price = df_clean['price'].max()
mean_price = df_clean['price'].mean()
median_price = df_clean['price'].median()

print("Minimum price:", min_price)
print("Maximum price:", max_price)
print("Mean price:", mean_price)
print("Median price:", median_price)
```
